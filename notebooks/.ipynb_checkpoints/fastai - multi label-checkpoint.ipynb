```python
def labels_as_string(data, row_array):
    a = data.columns.values[1:]
    b = row_array.iloc[1:]
    c = ' '.join(a[b > 0])
    return c
```

```python
data0 = data.copy()
data0.loc[:, 'tags'] = data.progress_apply(lambda x: labels_as_string(data, x), axis=1)
```
